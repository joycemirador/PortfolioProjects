### Import Libraries

```python
import json
import requests
import pandas as pd
```

### Establish Constants

```python
url = 'https://parseapi.back4app.com/classes/BR?limit=10000000'
headers = {
    'X-Parse-Application-Id': 'zS2XAEVEZAkD081UmEECFq22mAjIvX2IlTYaQfai',
    'X-Parse-Master-Key': 't6EjVCUOwutr1ruorlXNsH3Rz65g0jiVtbILtAYU'
}
csv_file_path = 'br_zipcodes.csv'
json_file_path = 'br_zipcodes.json'
```

### Access Data

```python
data = json.loads(requests.get(url, headers=headers).content.decode('utf-8'))
```

#### To easily access the data in the dictionary and convert it into a dataframe

```python
data_list = []
for item in data['results']:
    data_list.append(item)

for item in data_list:
    for x, y in item['geoPosition'].items():
        item[f'{x}'] = f'{y}'
    del item['geoPosition']

print(data_list[0])
```

```python
data_df = pd.DataFrame(data_list,
                       columns=['objectId', 'countryCode', 'postalCode', 'placeName',
                                'adminName1', 'adminCode1', 'adminName2', 'adminCode2',
                                'adminName3', 'adminCode3', 'accuracy', 'createdAt',
                                'createdAt', '__type', 'latitude', 'longitude'])
```

### Data Cleaning

```python
data_df.head()
```

```python
data_df.info()
```

```python
data_df['postalCode'] = data_df['postalCode'].str.replace('-', '')
data_df['postalCode'] = data_df['postalCode'].str.slice(stop=5)

data_df['postalCode'] = pd.to_numeric(data_df['postalCode'])
data_df['latitude'] = pd.to_numeric(data_df['latitude'])
data_df['longitude'] = pd.to_numeric(data_df['longitude'])

data_df[['postalCode', 'latitude', 'longitude']].info()
```

```python
data_df['placeName'] = data_df['placeName'].str.lower()
data_df.head()
```

### Save the Data into a CSV and a JSON file

The table import wizard in MySQL sometimes do not accept CSV files so it is best to save the data into both CSV and JSON files.

```python
data_cleaned = data_df[['postalCode', 'placeName', 'latitude', 'longitude']]
data_cleaned.to_csv(csv_file_path, index=False, encoding='utf-8')
```

```python
data_json = data_cleaned.to_json(orient='records', lines=False)
with open(json_file_path, 'w') as json_file:
  json_file.write(data_json)
```
